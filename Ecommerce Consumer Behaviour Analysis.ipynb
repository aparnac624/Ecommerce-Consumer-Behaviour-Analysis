### Fixing inconsistent datatypes and formatting 

```python
# fixing inconsistent datatypes
# fixing Purchase_Amount - converting to float
# Remove $ and , and whitespace, then convert to float
df['Purchase_Amount'] = df['Purchase_Amount'].astype(str)
df['Purchase_Amount'] = df['Purchase_Amount'].str.replace('$', '', regex=False)
df['Purchase_Amount'] = df['Purchase_Amount'].str.replace(',', '', regex=False)
df['Purchase_Amount'] = df['Purchase_Amount'].str.strip()
df['Purchase_Amount'] = df['Purchase_Amount'].astype(float).round(0)
# formatting values to 2 decimal places
df['Purchase_Amount'] = df['Purchase_Amount'].astype(int)
```

```python
# removing accents from column Location
import unicodedata

def remove_accents(text):
    if isinstance(text, str):
        return ''.join(
            c for c in unicodedata.normalize('NFKD', text)
            if not unicodedata.combining(c)
        )
    return text

df['Location'] = df['Location'].apply(remove_accents)
```

```python
import re

# Removing unwanted characters like weird quotes and backslashes
df['Location'] = df['Location'].astype(str).apply(lambda x: re.sub(r'[\"\'\\`’“”]', '', x).strip())
```

```python
# encoding multiclass categorical columns
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

# Define ordinal columns and order
ordinal_cols = {
    'Social_Media_Influence': ['None', 'Low', 'Medium', 'High'],
    'Engagement_with_Ads': ['None', 'Low', 'Medium', 'High']
}

# Map each category to its index
for col, order in ordinal_cols.items():
    order_dict = {k: v for v, k in enumerate(order)}
    df_encoded[col + '_Encoded'] = df_encoded[col].map(lambda x: order_dict.get(x, -1))  # -1 if unknown
```

```python
df_encoded
```

```python
# One-hot encoding nominal multiclass columns
nominal_cols = [
    'Gender', 'Marital_Status', 'Education_Level',
    'Purchase_Category', 'Purchase_Channel',
    'Discount_Sensitivity', 'Device_Used_for_Shopping',
    'Payment_Method', 'Shipping_Preference'
]

df_encoded = pd.get_dummies(df_encoded, columns=nominal_cols, prefix=nominal_cols, drop_first=True)
```

```python
df_encoded
```

### Handling Outliers

```python
# applying IQR method 
cols_with_outliers = ['Age', 'Frequency_of_Purchase', 'Time_Spent_on_Product_Research(hours)', 'Time_to_Decision', 'Customer_Satisfaction']

for col in cols_with_outliers:
    Q1 = df_encoded[col].quantile(0.25)
    Q3 = df_encoded[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df_encoded[(df_encoded[col] < lower) | (df_encoded[col] > upper)]
    print(f"{col} → Outliers: {len(outliers)}")
```
